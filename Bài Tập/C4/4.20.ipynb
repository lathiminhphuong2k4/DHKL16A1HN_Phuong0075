```python
import pandas as pd
```

```python
#Tạo data fame euro12
euro2012 = pd.read_csv('euro2012.csv')
print("Type of euro12:", type(euro2012))
print("Shape of euro12:", euro2012.shape)
print("Columns of euro12:", euro2012.columns)
```

```python
#in gt cột Goals
print("Goals column:")
print(euro2012['Goals'])
```

```python
#2 có bn đội tham gia Euro2012
num_teams=euro2012.shape[0]
print("Số đội tham gia Euro2012 là: ",num_teams)
```

```python
#3.IN thông tin của Euro2012
print("_____In thông tin Euro2012 _____")
print(euro2012.info())
```

```python
#4.Tạo 1 data frame mới từ euro2012 có tên là discipline chỉ chứa 3 cột 'Team', 'Yellow Cards', 'Red Cards'
discipline = euro2012[['Team', 'Yellow Cards', 'Red Cards']]
print("Data frame mới là: ")
discipline
```

```python
#5 Sắp xếp discipline giảm dần theo 2 cột 'Red Cards', 'Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("Sắp xếp discipline giảm dần theo hai cột là: ")
discipline_sorted
```

```python
#6a.Tính trung bình Yellow Cards
avg_yellow_cards = euro2012['Yellow Cards'].mean()
print("Trung bình Yellow Cards là: ",avg_yellow_cards)
```

```python
#6b.Lọc các đội đã ghi hơn 6 bàn thắng
doi_da_ghi_hon_6_ban_thang=euro2012[euro2012['Goals'] > 6]
print("Các đội đã ghi hơn 6 bàn thắng là: ")
doi_da_ghi_hon_6_ban_thang
```

```python
#7.In các đội mà tên bắt đầu bằng 'G'
teams_G=euro2012[euro2012['Team'].str.startswith('G')]
print("In các đội mà tên bắt đầu bằng 'G'")
teams_G
```

```python
#8 In 7 cột đầu của euro2012
print("In 7 cột đầu của euro2012:")
print(euro2012.iloc[:,:7])
```

```python
#9  In tất cả các cột, trừ 3 cột cuối
print(" In tất cả các cột, trừ 3 cột cuối:")
print(euro2012.iloc[:, :-3])
```

```python
#10 In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
selected_columns = ['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']
print("In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red :")
print(euro2012[selected_columns])
```

```python
#11 In các cột chi hiện thị 'Team','Shooting Accuracy' từ 'England', 'Italy', 'Russia'
selected_teams = ['England', 'Italy', 'Russia']
selected_columns = ['Team', 'Shooting Accuracy']
print("In các cột chi hiện thị 'Team','Shooting Accuracy' từ 'England',:")
print(euro2012[euro2012['Team'].isin(selected_teams)][selected_columns])
```
